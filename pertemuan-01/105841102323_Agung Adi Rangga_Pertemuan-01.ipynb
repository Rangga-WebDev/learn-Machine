# ***1️⃣ Array random 20 angka + mean, median, std***

```python
import numpy as np

data = np.random.randint(0, 101, 20)

mean= np.mean(data)
median = np.median(data)
std = np.std(data)

print("Array:", data)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)
```

# **2️⃣ DataFrame nilai 10 mahasiswa + 3 nilai tertinggi**

```python
import pandas as pd

data_nilai = {
    "Nama": [f"Mahasiswa_{i}" for i in range(1, 11)],
    "Nilai": np.random.randint(60, 100, 10)
}

df = pd.DataFrame(data_nilai)

print("Data Nilai:")
display(df)
topPriaSigma = df.sort_values(by="Nilai", ascending=False).head(3)

print("\n3 Nilai Tertinggi:")
display(topPriaSigma)
```

# ***3️⃣ Buat 2 grafik dari data nilai***

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
plt.bar(df["Nama"], df["Nilai"])
plt.xticks(rotation=45)
plt.title("Nilai Mahasiswa")
plt.xlabel("Nama")
plt.ylabel("Nilai")
plt.show()

plt.figure()
plt.hist(df["Nilai"], bins=5)
plt.title("Distribusi Nilai")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
plt.show()
```
